```python
#get products_info column values and insert into products_info table 
def products_(data, country):
    try:
        #get product part info
        products = data["products"]
        
        #change the type of object from list to dict
        products = products[0]
        
        #get upcList,eanList,categories
        try:
            upcList = products["upcList"]
            #get concact upclist for product_info table
            upcListCON = ""
            for upc in upcList:
                #concatenating the strings
                upcListCON = upc + "|" + upcListCON
            if upcListCON != "":
                upcListCON = upcListCON[:-1]
            upcListCON = codecs.escape_encode(str.encode(upcListCON))[0].decode()
        except KeyError as e:
            upcListCON = None
            
        try:
            eanList = products["eanList"]
            #get concact eanlist for product_info table
            eanListCON = ""
            for ean in eanList:
                #concatenating the strings
                eanListCON = ean + "|" + eanListCON
            if eanListCON != "":
                eanListCON = eanListCON[:-1]
            eanListCON = codecs.escape_encode(str.encode(eanListCON))[0].decode()
        except KeyError as e:
            eanListCON = None
            
        try:    
            categories = products["categories"]
             #get concact categories for product_info table
            categoriesCON = ""
            for category in categories:
                #concatenating the strings
                categoriesCON = str(category) + "|" + categoriesCON
            if categoriesCON != "":
                categoriesCON = categoriesCON[:-1]
            categoriesCON = codecs.escape_encode(str.encode(categoriesCON))[0].decode()
        except KeyError as e:
            categoriesCON = None
            
        
        try:
            frequently_bought_together = products["frequentlyBoughtTogether"]
            #get concact frequently_bought_together for product_info table
            frequently_bought_togetherCON = ""
            for frequentlyBought in frequently_bought_together:
                #concatenating the strings
                frequently_bought_togetherCON = frequentlyBought + "|" + frequently_bought_togetherCON
            if frequently_bought_togetherCON != "":
                frequently_bought_togetherCON = frequently_bought_togetherCON[:-1]
            frequently_bought_togetherCON = codecs.escape_encode(str.encode(frequently_bought_togetherCON))[0].decode()
        except KeyError as e:
            frequently_bought_togetherCON = None
        
        #get other values for product_info table
        try:
            parent_asin = products["parentAsin"]
        except KeyError as e:
            parent_asin = None
            
        try:    
            images = products["imagesCSV"]
            images = codecs.escape_encode(str.encode(images))[0].decode()
        except KeyError as e:
            images = None
            
        #print(parent_asin,images)    
        try:    
            title = products["title"]
            title = codecs.escape_encode(str.encode(title))[0].decode()
        except KeyError as e:
            title = None
            
        asin = products["asin"]
        product_type = products["productType"]
        country = country
        #timestamp = data["timestamp"]#in unix epoch time already, no conversion needed
            #upcListCON
            #eanListCON
            #categoriesCON
        tracking_since = (products["trackingSince"] + 21564000) * 60000
        last_update = (products["lastUpdate"] + 21564000) * 60000
        try:
            manufacturer = products["manufacturer"]
            manufacturer = codecs.escape_encode(str.encode(manufacturer))[0].decode()
        except KeyError as e:
            manufacturer = None
            
        root_category = products["rootCategory"]
        try:    
            brand = products["brand"]
            brand = codecs.escape_encode(str.encode(brand))[0].decode()
        except KeyError as e:
            brand = None
            
            #frequently_bought_togetherCON
        try:
            product_group = products["productGroup"]
            product_group = codecs.escape_encode(str.encode(product_group))[0].decode()
        except KeyError as e:
            product_group = None
            
        #insert values into products_info table 
        SQL_TABLE = "products"
        #parametrized version
        parameterized_stmt = "INSERT IGNORE INTO " + SQL_TABLE + " (title,\
                                                asin,\
                                                parent_asin,\
                                                images,\
                                                product_type,\
                                                country,\
                                                upcs,\
                                                eans,\
                                                categories,\
                                                tracking_since,\
                                                manufacturer,\
                                                root_category,\
                                                last_update,\
                                                brand,\
                                                frequently_bought_together,\
                                                product_group) values (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s,%s, %s, %s, %s, %s);"

        cursor.execute(parameterized_stmt, (title, 
                            asin,
                            parent_asin,
                            images,
                            product_type,
                            country,
                            upcListCON,
                            eanListCON,
                            categoriesCON,
                            tracking_since,
                            manufacturer,
                            root_category,
                            last_update,
                            brand,
                            frequently_bought_togetherCON,
                            product_group))
        conn.commit()
        return

    except IOError as e:
        print(e)
```

```python
#get amazon price histroy and insert into amazon_price table 
def prices_amazon(csv, country, asin):
    try:
        try:
            #get amazon price histroy
            amazon_price = csv[0]
            #insert values into amazon_price table 
            SQL_TABLE = "prices_amazon"
            #standard version
            prepared_stmt = "INSERT IGNORE INTO " + SQL_TABLE + " (asin, time, price, country) values ('{}', '{}', '{}', '{}')"
            #amazon_price_insert = []
            for i in range(0,len(amazon_price)-1,2):
                time = (int(amazon_price[i]) + 21564000) * 60000
                price = int(amazon_price[i+1])
                country = country
                stmt = prepared_stmt.format(asin,
                                time,
                                price,
                                country)
                #print(stmt)
                cursor.execute(stmt)
            conn.commit()
        except TypeError as te:
            print("no amazon price info")
        return

    except IOError as e:
        print(e)
```

```python
#get frequently_bought_together values and insert into frequently_Bought_Together table 
def frequently_bought_together(products, country):
    try:
        #get frequently_Bought_Together histroy
        try:
            frequently_bought_together = products["frequentlyBoughtTogether"]
        
            #insert frequently_Bought_Together values
            SQL_TABLE = "frequently_bought_together"
            #standard version
            prepared_stmt = "INSERT IGNORE INTO " + SQL_TABLE + " (asin, bought_together_asin, country) values ('{}', '{}', '{}')"
            for i in range(0,len(frequently_bought_together)):
                asin = products["asin"]
                btasin = frequently_bought_together[i]
                country = country
                stmt = prepared_stmt.format(asin,
                                btasin,
                                country)
                #print(stmt)
                cursor.execute(stmt)
            conn.commit()
        except KeyError as e:
            print("no frequently bought together info")
        
        return

    except IOError as e:
        print(e)
```

```python
#get variations values and insert into variations table 
def variations(products, country):
    try:
        variations = products["variations"]
        SQL_TABLE = "variations"
        #standard version
        prepared_stmt = "INSERT IGNORE INTO " + SQL_TABLE + " (asin, variation_asin, dimension, value, country) values ('{}', '{}', '{}', '{}', '{}')"
        for i in range(0, len(variations)):
            #print(i)
            asin = products["asin"]
            variationsAsin = variations[i]["asin"]
            #print(variations[i]["attributes"][0])
            dimension = variations[i]["attributes"][0]["dimension"]
            value = variations[i]["attributes"][0]["value"]
            country = country
            stmt = prepared_stmt.format(asin,
                                variationsAsin,
                                codecs.escape_encode(str.encode(dimension))[0].decode(),
                                codecs.escape_encode(str.encode(value))[0].decode(),
                                country)
            #print(stmt)
            cursor.execute(stmt)
        conn.commit()
    
    except KeyError as e:
        print("no variations information")
    
        return

    except IOError as e:
        print(e)
```
